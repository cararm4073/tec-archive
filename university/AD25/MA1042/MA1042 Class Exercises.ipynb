#Exercise 12

* Do a recursive function `sierpinski(size, level)` that draws a sierpinski curve of a given size and level as discussed in class:
  * level zero is a triangle
  * And as the level progresses, as discussed in class the picture looks more and more.

The code must be similar to the tree example, and should not be iterative or partly iterative.

If one of your classmates did the code, give credit.

**Do not ask chat-gpt or similar**

```python
import ColabTurtle.Turtle as jace
jace.initializeTurtle()
jace.clear()

def sierpinski(size: int,level:int):
  jace.speed(5)

  if level == 0: #for a single triangle
    for i in range(3):
      jace.forward(size)
      jace.right(120)

  else:
    sierpinski(size / 2, level - 1) #left lower triangle
    jace.forward(size / 2)

    sierpinski(size / 2, level - 1) #right lower triangle
    jace.right(120)
    jace.forward(size / 2)
    jace.left(120)

    sierpinski(size / 2, level - 1) #top triangle
    jace.left(120)
    jace.forward(size / 2)
    jace.right(120)

jace.clear()
sierpinski(size=250, level=3)
```
